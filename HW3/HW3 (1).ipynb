```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

# 定義要爬取的網址
url = 'https://www.etax.nat.gov.tw/etw-main/ETW183W2_11301/'

# 發送HTTP請求，獲取網頁內容
response = requests.get(url)
response.encoding = 'utf-8'  # 指定編碼方式為UTF-8
soup = BeautifulSoup(response.text, 'html.parser')

# 使用正則表達式找到所需的資料
# 這裡以抓取所有<td,br,th>標籤內的文字
td_tags = soup.find_all('td')
data1 = [td.get_text(strip=True) for td in td_tags]
br_tags = soup.find_all('br')
data2 = [br.get_text(strip=True) for br in br_tags]
th_tags = soup.find_all('th')
data3 = [th.get_text(strip=True) for th in th_tags]

# 找到所有的文字內容
text_elements = soup.find_all(['td', 'br', 'th'])
text_data = [element.get_text(strip=True) for element in text_elements]

# 找到所有的表格內容
table_elements = soup.find_all('tr')
table_data = []
for row in table_elements:
    row_data = [cell.get_text(strip=True) for cell in row.find_all(['td', 'th'])]
    table_data.append(row_data)


# 將文字內容結構化為DataFrame
text_df = pd.DataFrame({'text_data': text_data})

# 將表格內容結構化為DataFrame
table_df = pd.DataFrame(table_data[1:], columns=table_data[0])

# 將文字內容結構化為字典
text_dict = {'text_data': text_data}

# 將表格內容結構化為字典
table_dict = table_df.to_dict(orient='records')

# 將td內容遇到逗號就換行
table_df["113年 01 ~ 02 月"] = table_df["113年 01 ~ 02 月"].str.replace(",", "\n")





# 輸出DataFrame為CSV檔案
table_df.to_csv('data.csv', index=False)

# 輸出DataFrame為JSON檔案
table_df.to_json('data.json', orient='records', force_ascii=False, indent=4)


# 輸出結果
print("\n表格內容DataFrame:")
print(table_df.head())
print("\n表格內容dict:")
print(table_dict)
print('資料已成功輸出為CSV和JSON檔案')




```
