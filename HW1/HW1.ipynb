```python
import pandas as pd

# 讀取 CSV 檔案
df = pd.read_csv('hw1 - 工作表1.csv')

# 篩選出 Growing Zone 為 Tropical 的資料
tropical_fruits = df[df['Growing Zone'] == 'Tropical']

# 篩選出 Red Color 為 Yes 的資料
red_fruits = df[df['Red Color'] == 'Yes']

# 將兩個篩選結果合併為一個資料集(聯集)
tropical_and_red_fruits = pd.concat([tropical_fruits, red_fruits])

# 輸出統整結果
print(tropical_and_red_fruits)

# 篩選出 Growing Zone 為 Tropical 且 Red Color 為 Yes 的資料(交集)
tropical_red_fruits = df[(df['Growing Zone'] == 'Tropical') & (df['Red Color'] == 'Yes')]

# 輸出篩選結果
print(tropical_red_fruits)

# 篩選出 Growing Zone 為 Tropical 且 Red Color 為 No 的資料(差集)
tropical_non_red_fruits = df[(df['Growing Zone'] == 'Tropical') & (df['Red Color'] == 'No')]

# 輸出篩選結果
print(tropical_non_red_fruits)

# 篩選出 Growing Zone 不是 Tropical 但 Red Color 是 Yes 的資料(差集)
non_tropical_red_fruits = df[(df['Growing Zone'] != 'Tropical') & (df['Red Color'] == 'Yes')]

# 輸出統整結果
print(non_tropical_red_fruits)
```
